# Retraining

## Second run (small)

```python
model = YOLO("/home/student/Synthetic-data/Part2/runs/pose/v8-small-add-data/weights/best.pt")  

model.train(
    data="pseudo_data.yaml",
    imgsz=960,
    epochs=300,
    patience = 70,
    batch=16,
    workers=8,
    # lr0=0.005,
    project="runs/pose",
    name="pseudo_synth",
    augment=True,
    # geometric
    degrees=5.0,
    translate=0.30,
    scale=0.40,
    shear=0.0,
    perspective=0.0005,
    flipud=0.30,
    fliplr=0.50,

    # color/intensity
    hsv_h=0.005,
    hsv_s=0.30,
    hsv_v=0.45,

    # mix-based
    mosaic=0.20,              # prob of mosaic
    mixup=0.10,               # prob of mixup
    copy_paste=0.20,          # prob of copy-paste (useful for occlusions)             
)
```
